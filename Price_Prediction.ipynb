```python
df["name"]=df["name"].apply(lambda x: x.replace(x,"") if "sector" in x else x)
```

```python
i=-1
while True:
    i=i+1
    cityName.append(df["location"][i]["cityName"])
    addressLocality.append(df["location"][i]["addressLocality"])
    longitude.append(df["location"][i]["longitude"])
    latitude.append(df["location"][i]["latitude"])
    if i==57422:
        break
```
